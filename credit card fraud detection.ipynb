Importing Libraries

```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Import Dataset
(V1-V28 are confidential details which are converted by PCA already ) 

```python
df=pd.read_csv('creditcard.csv')
```

```python
#separating data for analysis into fraud and legit

legit=df[df['Class']==0]
legit
```

Under Sampling
(Build a sample dataset containing similar distribution of legit and fraud)

no. of fraud=492

```python
legit_sample= legit.sample(n=492)
```

Concatenating two Dataframes

```python
new_df=pd.concat([legit_sample,fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

Splitting data into Features and Target

```python
X=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
```

```python
print(X)
```

```python
print(y)
```

Split into Training and Testing

```python
X_train,X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, stratify=y, random_state=2)
```

```python
X_train
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training 

```python
model =LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
test_predict=model.predict(X_test)
```

Evaluating the Model

```python
#first test accuracy on training 
train_predict=model.predict(X_train)
training_accuracy=accuracy_score(train_predict, y_train)
```

```python
print("Accuracy on Training Data: ", training_accuracy)
```

```python
test_accuracy=accuracy_score(test_predict,y_test)
print("Accuracy on Testing Data: ", test_accuracy)
```
